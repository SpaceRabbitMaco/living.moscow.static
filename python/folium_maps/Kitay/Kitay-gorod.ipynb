```python
import folium
from folium import plugins
from folium.plugins import MiniMap

import sys
sys.path.append('../../tools')

import MapTools as mt
```

```python
# Define vars

attractions = mt.load_markers('./static_data/markers/attractions/')
museums = mt.load_markers('./static_data/markers/museums/')
parks = mt.load_markers('./static_data/markers/parks/')
wcs = mt.load_markers('./static_data/markers/wcs/')

areas = mt.load_areas('./static_data/areas/')

names = ['Китай-город', 'Парк Зарядье']

styles = [{'fillColor':'#223e8b','color':'#223e8b','opacity':0.5},
          {'fillColor': '#228B22', 'color': '#228B22', 'opacity': 0.3}]
```

```python
# Set Kitay-gorod start location
m = folium.Map(location=[55.753006, 37.625473],
    tiles='Stamen Toner',
    zoom_start=15)

# Add areas to Map   
mt.add_areas_to_map(areas, names, styles, m)

# Add minimap to Map
minimap = MiniMap(tile_layer='Stamen Toner')
m.add_child(minimap)

# Add markers on groups
fg = folium.FeatureGroup(control=False) # Main group
g1 = folium.plugins.FeatureGroupSubGroup(fg, 'Интересные места')
g2 = folium.plugins.FeatureGroupSubGroup(fg, 'Музеи, Театры, Выставки')
g3 = folium.plugins.FeatureGroupSubGroup(fg, 'Парки')
g4 = folium.plugins.FeatureGroupSubGroup(fg, 'Туалеты (WC)')
m.add_child(fg)
m.add_child(g1)
m.add_child(g2)
m.add_child(g3)
m.add_child(g4)

# Add markers on Map
mt.add_markers_to_map(attractions, g1)
mt.add_markers_to_map(museums, g2)
mt.add_markers_to_map(parks, g3)
mt.add_markers_to_map(wcs, g4)

# Add Layer Control & Tools
folium.LayerControl().add_to(m)
plugins.Fullscreen(position='topright').add_to(m)

m
```
