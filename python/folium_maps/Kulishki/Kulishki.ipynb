```python
import folium
from folium import plugins
from folium.plugins import MiniMap

import sys
sys.path.append('../../tools')

import MapTools as mt
```

```python
# Define vars

areas = mt.load_areas('./static_data/areas/')

attractions = mt.load_markers('./static_data/markers/attractions/')
parks = mt.load_markers('./static_data/markers/parks/')
business = mt.load_markers('./static_data/markers/business/')
museums = mt.load_markers('./static_data/markers/museums/')

names = ['Кулишки (Ивановская горка)',]

styles = [{'fillColor':'#223e8b','color':'#223e8b','opacity':0.5},{'fillColor':'#228b82','color':'#228b82','opacity':0.5},
         {'fillColor':'#228B22','color':'#228B22','opacity':0.5},{'fillColor':'#8b2222','color':'#8b2222','opacity':0.5},
         {'fillColor':'#228b82','color':'#228b82','opacity':0.5},{'fillColor':'#8b2282','color':'#8b2282','opacity':0.5},
         {'fillColor':'#50228b','color':'#50228b','opacity':0.5},{'fillColor':'#8b5c22','color':'#8b5c22','opacity':0.5},
         {'fillColor':'#aba505','color':'#aba505','opacity':0.5},{'fillColor':'#228B22','color':'#228B22','opacity':0.5}]
```

```python
# Set Kulishki start location
m = folium.Map(location=[55.757006, 37.625473],
    tiles='Stamen Toner',
    zoom_start=14)

# Add areas to Map   
mt.add_areas(areas, names, styles, m)

# Add minimap to Map
minimap = MiniMap(tile_layer='Stamen Toner')
m.add_child(minimap)

# Add markers on groups
fg = folium.FeatureGroup(control=False)                           # Main group
g1 = folium.plugins.FeatureGroupSubGroup(fg, 'Интересные места')  # First subgroup of fg
g2 = folium.plugins.FeatureGroupSubGroup(fg, 'Парки')  # First subgroup of fg
g3 = folium.plugins.FeatureGroupSubGroup(fg, 'Кафе и Магазины')  # First subgroup of fg
g4 = folium.plugins.FeatureGroupSubGroup(fg, 'Музеи, Театры, Выставки')  # First subgroup of fg

m.add_child(fg)
m.add_child(g1)
m.add_child(g2)
m.add_child(g3)
m.add_child(g4)

# Add markers on Map
mt.add_markers_to_map(attractions, g1)
mt.add_markers_to_map(parks, g2)
mt.add_markers_to_map(business, g3, True)
mt.add_markers_to_map(museums, g4, True)

# Add Layer Control & Tools
folium.LayerControl().add_to(m)
plugins.Fullscreen(position='topright').add_to(m)

m
```
